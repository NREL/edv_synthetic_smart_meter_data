```python
import pandas as pd
import plotly.graph_objects as go
import numpy as np
```

where is the csv file?

```python
file = 'CBES_BRICR_baselines.xlsx'
```

how does the results file look like?

```python
df = pd.read_excel(file)
df
```

how are baseline results look compared to actual monthly billing?

```python
jan = go.Scatter(x=df.utility_elec_m_1.to_list(), y=df.baseline_elec_m_1.to_list(), mode='markers', name='JAN', text = df.seed_primary_property_type)
feb = go.Scatter(x=df.utility_elec_m_2.to_list(), y=df.baseline_elec_m_2.to_list(), mode='markers', name='FEB', text = df.seed_primary_property_type)
mar = go.Scatter(x=df.utility_elec_m_3.to_list(), y=df.baseline_elec_m_3.to_list(), mode='markers', name='MAR', text = df.seed_primary_property_type)
apr = go.Scatter(x=df.utility_elec_m_4.to_list(), y=df.baseline_elec_m_4.to_list(), mode='markers', name='APR', text = df.seed_primary_property_type)
may = go.Scatter(x=df.utility_elec_m_5.to_list(), y=df.baseline_elec_m_5.to_list(), mode='markers', name='MAY', text = df.seed_primary_property_type)
jun = go.Scatter(x=df.utility_elec_m_6.to_list(), y=df.baseline_elec_m_6.to_list(), mode='markers', name='JUN', text = df.seed_primary_property_type)
jul = go.Scatter(x=df.utility_elec_m_7.to_list(), y=df.baseline_elec_m_7.to_list(), mode='markers', name='JUL', text = df.seed_primary_property_type)
aug = go.Scatter(x=df.utility_elec_m_8.to_list(), y=df.baseline_elec_m_8.to_list(), mode='markers', name='AUG', text = df.seed_primary_property_type)
sep = go.Scatter(x=df.utility_elec_m_9.to_list(), y=df.baseline_elec_m_9.to_list(), mode='markers', name='SEP', text = df.seed_primary_property_type)
oct = go.Scatter(x=df.utility_elec_m_10.to_list(), y=df.baseline_elec_m_10.to_list(), mode='markers', name='OCT', text = df.seed_primary_property_type)
nov = go.Scatter(x=df.utility_elec_m_11.to_list(), y=df.baseline_elec_m_11.to_list(), mode='markers', name='NOV', text = df.seed_primary_property_type)
dec = go.Scatter(x=df.utility_elec_m_12.to_list(), y=df.baseline_elec_m_12.to_list(), mode='markers', name='DEC', text = df.seed_primary_property_type)

data = [jan, feb, mar, apr, may, jun, jul, aug, sep, oct, nov, dec]

fig = go.Figure(data=data)

fig.update_layout(
    title="Monthly parity plot: Electricity (before calibration)",
    xaxis=dict(
        title='Actual monthly consumptions [kWh]',
        range=[0,200000],
    ),
    yaxis=dict(
        title='Simulated monthly consumptions [kWh]',
        range=[0,200000],
    ),
    width=800,
    height=800,
)

fig.show()
```
