```python
import pandas as pd
import plotly.graph_objects as go
import numpy as np
import plotly.express as px
import os
```

where is the csv file?

```python
file = 'CBES_BRICR_baselines.xlsx'
```

how does the results file look like?

```python
df = pd.read_excel(file)
df
```

```python
for col in df.columns:
    print(col)
```

```python
df = df.rename(columns={
    "seed_primary_property_type": "Building Type", 
    "B": "c"
})
```

```python
df['utility_gas_annual'] = 0
df['utility_elec_annual'] = 0
for num in range(12):
    df['utility_gas_annual'] = df['utility_gas_annual'] + df['utility_gas_m_{}'.format(num+1)]
    df['utility_elec_annual'] = df['utility_elec_annual'] + df['utility_elec_m_{}'.format(num+1)]
    
df
```

```python
df['EUI_elec'] = df['utility_elec_annual'] / df['seed_property_gfa_m2']
df['EUI_gas'] = df['utility_gas_annual'] / df['seed_property_gfa_m2']
df
```

```python
df['Count'] = 1
df['All'] = 'All'
```

```python
# plot setting
stn_plot_bgcolor = 'rgba(0,0,0,0)'
stn_plot_linecolor = 'rgba(0,0,0,0.5)'
stn_histogram_opacity = 0.75
stn_title_font = 12
stn_width = 250
stn_height = 400
stn_margin = dict(
    l=50,
    r=50,
    t=50,
    b=50
)
stn_marginaltype = 'box'
stn_annotation_x = 0.4
stn_annotation_y = 1
stn_plot_maincolor = 'royalblue'
```

```python
df.seed_property_gfa_m2.max()
```

```python
avg = round(df.seed_property_gfa_m2.mean())
med = round(df.seed_property_gfa_m2.median())

fig = px.histogram(
    df, 
    y="seed_property_gfa_m2", 
#     x="Count", 
    color="All",
    marginal=stn_marginaltype, # or violin, rug
    nbins=100,
    color_discrete_sequence=[stn_plot_maincolor]
)

fig.update_layout(
#     barmode='overlay',
    width=stn_width,
    height=stn_height,
    yaxis = dict(
        title='<b>Gross Building Floor Area [m<sup>2</sup>]</b>',
        title_font = {"size": stn_title_font},
        showgrid=False,
    ),
    xaxis = dict(
        title='<b>Number of Buildings</b>',
        title_font = {"size": stn_title_font},
        showgrid=False,
    ),
    margin=stn_margin,
    plot_bgcolor=stn_plot_bgcolor,
    showlegend=False,
)

fig.add_annotation(
    x=stn_annotation_x,
    y=stn_annotation_y,
    xref="paper",
    yref="paper",
    text="Average = {}<br>Median = {}".format(avg, med),
    showarrow=False,
    font=dict(
        size=10,
        color="black"
        ),
    align="left",
    borderwidth=1,
    borderpad=4,
    opacity=0.8
)

fig.update_xaxes(
    mirror=True,
    showline=True, 
    linewidth=1, 
    linecolor=stn_plot_linecolor
)

fig.update_yaxes(
    mirror=True,
    showline=True, 
    linewidth=1, 
    linecolor=stn_plot_linecolor
)

fig.update_layout(legend=dict(
#     yanchor="top",
#     y=0.99,
#     xanchor="left",
#     x=0.05,
    font=dict(
        size=9,
        color="black"
    ),
    title=dict(
        font=dict(
            size=10,
            color="black"
        ),
    ),
))

fig.update_traces(opacity=stn_histogram_opacity)

fig.show()

if not os.path.exists("images"):
    os.mkdir("images")

fig.write_image("images/fig_{}.png".format("GrossFloorArea"))
```

```python
fig = go.Figure()

df_max = df.filter(like='utility_elec_m_', axis=1).max(axis=1)

i=1
for month in ['JAN','FEB','MAR','APR','MAY','JUN','JUL','AUG','SEP','OCT','NOV','DEC']:

    fig.add_trace(go.Box(
        x=df['utility_elec_m_{}'.format(i)]/df_max, 
        name=month,
        marker_color = stn_plot_maincolor
    ))
    
    i+=1

fig.update_layout(
    width=stn_width,
    height=stn_height,
    yaxis = dict(
#         title='Site Natural Gas EUI [kWh/m<sup>2</sup>]',
#         title_font = {"size": stn_title_font},
        showgrid=False,
    ),
    xaxis = dict(
        title='<b>Normalized Electricity Usage</b>',
        title_font = {"size": stn_title_font},
        showgrid=False,
    ),
    margin=stn_margin,
    plot_bgcolor=stn_plot_bgcolor,
    showlegend=False,
)

fig.update_xaxes(
    mirror=True,
    showline=True, 
    linewidth=1, 
    linecolor=stn_plot_linecolor
)

fig.update_yaxes(
    mirror=True,
    showline=True, 
    linewidth=1, 
    linecolor=stn_plot_linecolor
)

fig.show()

if not os.path.exists("images"):
    os.mkdir("images")

fig.write_image("images/fig_{}.png".format("MonthlyNorm_Elec"))
```
